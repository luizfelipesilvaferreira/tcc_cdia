```python
import pandas as pd
df = pd.read_csv('df_final_cpa.csv')
df
```

```python
# Calculando a frequência e adicionando a coluna FREQUÊNCIA_CP
frequencia_cp = df.groupby(['MUNICÍPIO', 'CNAE_FISCAL_PRINCIPAL']).size().reset_index(name='FREQUÊNCIA_CP')
df = pd.merge(df, frequencia_cp, on=['MUNICÍPIO', 'CNAE_FISCAL_PRINCIPAL'], how='left')
df
```

```python
# efetuando as transformações necessárias nos dados:
df['MUNICÍPIO'] = df['MUNICÍPIO'].astype(str)
```

```python
def top_10_cnaes_por_municipio(df, municipio):
    # Filtrar o DataFrame apenas para o MUNICÍPIO fornecido como entrada
    df_municipio = df[df['MUNICÍPIO'] == municipio]
    
    # Calcular a frequência de CNAE_FISCAL_PRINCIPAL para o município
    frequencia_cp = df_municipio['CNAE_FISCAL_PRINCIPAL'].value_counts().reset_index()
    frequencia_cp.columns = ['CNAE_FISCAL_PRINCIPAL', 'FREQUÊNCIA_CP']
    
    # Ordenar os resultados em ordem decrescente de frequência
    frequencia_cp = frequencia_cp.sort_values('FREQUÊNCIA_CP', ascending=False)
    
    # Selecionar os top 10 códigos CNAE_FISCAL_PRINCIPAL com maiores valores de FREQUÊNCIA_CP
    top_10_cnaes = frequencia_cp.head(10)
    
    return top_10_cnaes

# Exemplo de uso
municipio_desejado = '8569'  # Substitua pelo nome do município desejado
top_cnaes_municipio = top_10_cnaes_por_municipio(df, municipio_desejado)

print(f"Top 10 CNAE_FISCAL_PRINCIPAL com maior FREQUÊNCIA_CP em {municipio_desejado}:")
top_cnaes_municipio
```
