```python
import pandas as pd

# Carregar os DataFrames dos arquivos CSV
df0_cpa = pd.read_csv('df0_cpa.csv')
df1_cpa = pd.read_csv('df1_cpa.csv')
df2_cpa = pd.read_csv('df2_cpa.csv')
df3_cpa = pd.read_csv('df3_cpa.csv')
df4_cpa = pd.read_csv('df4_cpa.csv')
df5_cpa = pd.read_csv('df5_cpa.csv')
df6_cpa = pd.read_csv('df6_cpa.csv')
df7_cpa = pd.read_csv('df7_cpa.csv')
df8_cpa = pd.read_csv('df8_cpa.csv')
df9_cpa = pd.read_csv('df9_cpa.csv')

# Concatenar os DataFrames verticalmente
df_unificado = pd.concat([df0_cpa, df1_cpa, df2_cpa, df3_cpa, df4_cpa, df5_cpa, df6_cpa, df7_cpa, df8_cpa, df9_cpa])

df_unificado['DATA_DE_INÍCIO_ATIVIDADE'] = df_unificado['DATA_DE_INÍCIO_ATIVIDADE'].astype(str).str[:4]
df_unificado['DATA_DE_INÍCIO_ATIVIDADE'] = df_unificado['DATA_DE_INÍCIO_ATIVIDADE'].astype(int)
df_unificado = df_unificado.sort_values(by='DATA_DE_INÍCIO_ATIVIDADE', ascending=False)
df_unificado
```

```python
df = df_unificado
# Filtrar registros com erro no ano
registros_com_erro = df[df['DATA_DE_INÍCIO_ATIVIDADE'] < 1694]

# Exibir a quantidade de valores com erro no ano de registro
quantidade_erros = len(registros_com_erro)
print("Quantidade de valores com erro no ano de registro:", quantidade_erros)

# Remover os registros com erro do DataFrame
df = df[df['DATA_DE_INÍCIO_ATIVIDADE'] >= 1694]
df
```

```python
# salvando em .csv
df = df.to_csv('df_final_cpa.csv', index=False)
```
