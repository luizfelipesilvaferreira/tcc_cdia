```python
import pandas as pd

# Criação do DataFrame de exemplo
df = pd.read_csv('df9.csv')
#df = pd.DataFrame('df0.csv')
df['CNAE_FISCAL_PRINCIPAL'] = df['CNAE_FISCAL_PRINCIPAL'].astype(str)
# Adição da coluna com o número de caracteres
df['num_caracteres'] = df['CNAE_FISCAL_PRINCIPAL'].str.len()
```

```python
# Ordenação em ordem decrescente
df = df.sort_values('num_caracteres', ascending=False)
```
